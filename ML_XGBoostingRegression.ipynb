```python
#Getting library files into our program
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#gettig data from client sheet
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#remove from duplicates from client data set
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#display data post removal of dummies
dataset
```

```python
#display columns post removal of dummies
dataset.columns
```

```python
#assign the input values to independent, output values to dependent variable
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
#display the input values
independent
```

```python
#display output values
dependent
```

```python
#now the model selection and splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#display training data
X_train
```

```python
#display test set data
X_test
```

```python
#display output training data
y_train
```

```python
#display output testset data
y_test
```

```python
#now getting the regression model here
from sklearn.ensemble import GradientBoostingRegressor
regressor = GradientBoostingRegressor()
regressor = regressor.fit(X_train, y_train)
```

```python
#display regressor model
regressor
```

```python
#predict testset data
y_pred = regressor.predict(X_test)
```

```python
#now importing here the evaluation metrics
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
#display the r_score
r_score
```

```python
#getting pickle module for deployment phase
import pickle
```

```python
#saving the file model to filename variable
filename = "finalized-model_XGboostRegressor_Assignment_DafultValues.sav"
```

```python
#usimg pickle dump method we are writing the file on disk
pickle.dump(regressor, open(filename, 'wb'))
```
