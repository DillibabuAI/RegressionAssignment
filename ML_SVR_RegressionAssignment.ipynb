```python
#Getting library files
import pandas as pd
```

```python
#Storing client data values in the variable named "dataset"
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#Displaying the client data stored
dataset
```

```python
#Displaying the columns header names from client data sheet
dataset.columns
```

```python
#Removing duplicates from data set using data preprocessing method
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#Displaying dataset again post removing duplicates
dataset
```

```python
#Displaying dataset row, columns count
dataset.shape
```

```python
#Displaying column names post data preprocessing.
dataset.columns
```

```python
#Stroing input data values to independent variable
independent = dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
#Displaying independent variable data
independent
```

```python
#Storing output data values in dependent variable
dependent = dataset[['charges']]
```

```python
#Displaying dependent variable stored data
dependent
```

```python
#Using sklearn selecting the model and splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#Displaying the training set data
X_train
```

```python
#Displaying the test set data
X_test
```

```python
#now creating the model using sklearn
from sklearn.svm import SVR
regressor = SVR(kernel = 'linear', C=3000)
```

```python
#assigning the model to regressor variable
regressor.fit(X_train, y_train)
```

```python
#getting slope value assigned to weight variable
weight = regressor.intercept_
```

```python
#displaying the intercept value
weight
```

```python
#displaying the length
regressor.n_support_
```

```python
#Storing test set data predicted values in the variable y_pred
y_pred = regressor.predict(X_test)
```

```python
#Getting metrics, r2_score from library
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
#Displaying r_score value
r_score
```

```python
#Getting pickle library for deployment phase
import pickle
```

```python
#giving file name to out model
filename = "finalized_SVM_Linear_C3000.sav"
#using pickle dump method we are writing this file on disk
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#loading the saved model into loaded_model variable
loaded_model = pickle.load(open("finalized_SVM_Linear_C3000.sav", 'rb'))
```

```python
#here we are pedicting the value like we want and assigning to result variable.
result = loaded_model.predict([[46,786.56,1,True,False]])
```
