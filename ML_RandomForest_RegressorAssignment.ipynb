```python
#getting library files
import pandas as pd
```

```python
#stroing the client data into dataset variable
dataset = pd.read_csv("insurance_pre.csv")
```

```python
#displaying the dataset
dataset
```

```python
#display number of rows, columns
dataset.shape
```

```python
#removing dummies from dataset
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#display dataset post removing dummies
dataset
```

```python
#display column names of dataset
dataset.columns
```

```python
#assign the input data to independent variable
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
#assign the output data to dependent variable
dependent = dataset[['charges']]
```

```python
#display input data shape and column values
independent
```

```python
#displaying the output data values
dependent
```

```python
#here we are doing model_selection and splitting the training, test set data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#display the training data set values
X_train
```

```python
#display x_test data
X_test
```

```python
#display the shape of training data
X_train.shape
```

```python
#display the shape of test set data
X_test.shape
```

```python
#from sklearn ensemble module we are importing RandomForestRegressor and assiging the method to regressor variable
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(criterion = 'poisson', n_estimators = 50)
regressor.fit(X_train, y_train)
```

```python
#predicting the value and assigning to y_pred variable
y_pred = regressor.predict(X_test)
```

```python
#importing evaluation metrics and calculate r_score
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
#display the r_score value
r_score
```

```python
#import pickle library for deployment phase
import pickle
```

```python
#assign the saving model file name to filename variable
filename = "finalized-model_RandomForestRegressor_Poisson50.sav"
#using pickle dump method write the file on disk
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#loading the saved model to loaded_variable.
loaded_model = pickle.load(open(filename, 'rb'))
```

```python
#here we are pedicting the value like we want and assigning to result variable.
result = loaded_model.predict([[21, 432.34, 2, True, False]])
```
